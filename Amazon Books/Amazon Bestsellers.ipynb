# Amazon Bestsellers

```python
# Importing libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
# uploading data
data = pd.read_csv('bestsellers with categories.csv')
data.head()
```

```python
data.info()
```

- No null objects

```python
data.describe()
```

```python
data.Author.value_counts()
```

```python
fig , (axis1, axis2) = plt.subplots(2,1,figsize = (10,10))
sns.countplot(x='Author', data=data, ax=axis1)
sns.countplot(x='Genre', data=data, ax=axis2)
```

```python
# count of fiction and non ficton books
data.groupby('Genre').count()
```

```python
# maximum user ratings for different genres
data.groupby('Genre')['User Rating'].max().plot(kind = 'bar', figsize=(10,6), color = 'slategrey')
plt.title('Max user rating for different genre')
plt.xlabel('Genre')
plt.ylabel('User Rating')
plt.show()
```

- Ficyion and non fiction ratings is same

```python
# Minimum user ratings for different genre
data.groupby('Genre')['User Rating'].min().plot(kind = 'bar', figsize=(10,6), color = 'lightsteelblue')
plt.title('Min user rating for different genre')
plt.xlabel('Genre')
plt.ylabel('User Rating')
plt.show()
```

- For fiction the minimum rating is 3.5 and for non fiction, minimum e=rating is 4

```python
# top 10 authors with highest rated books

data.groupby('Author')['User Rating'].max().sort_values(ascending = False).head(10).plot(kind = 'bar', figsize = (12,6), color = 'cornflowerblue')
plt.title('Top 10 authors with the highest ratings')
```

```python
# Top 10 books with highest ratings
data.groupby('Name')['User Rating'].max().sort_values(ascending = False).head(10).plot(kind = 'bar', figsize = (12,6), color = 'royalblue')
plt.title('Top 10 books with highest ratings')
```

```python
# Top 10 boks with highest reviews
data.groupby('Name')['Reviews'].max().sort_values(ascending = False).head(10).plot(kind = 'bar', figsize = (12,6), color = 'royalblue')
plt.title('Top 10 books with highest reviews')
```

```python
# Heatmap
plt.figure(figsize = (10,5))
sns.heatmap(data.corr(), annot = True)
```

```python
# Percentage of fiction and non fiction books
color = ['teal', 'cyan']
data['Genre'].value_counts().plot(kind = 'pie', figsize = (8,8), autopct = '%1.2f', colors = color, explode = [0,0.02])
plt.axis = ('equal')
plt.legend(loc = 'upper right')
plt.show()
```
